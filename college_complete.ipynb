## Assignment

- Create a for loop that plots the histogram of sample size, median, employed, full time, unemployment rate, men and women (remember to use the exact column titles)

```python
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize = (5,12))

for i in range(1,5):
    ax = fig.add_subplot(4,1,i)
    ax = grads[cols[i-1]].plot(kind="hist")

plt.show()
```

```python
fig = plt.figure(figsize = (5,12))

for i in range(4,7):
    ax = fig.add_subplot(4,1,i-3)
    ax = grads[cols[i-1]].plot(kind="hist")

plt.show()
```
